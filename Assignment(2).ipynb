### 11.	Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then:
- Use .info(), .describe()
- Select 'Name' and 'City' columns
- Drop 'City' column
- Fill any missing values in 'Salary' column with the mean
- Remove any duplicate rows


```python
import pandas as pd
import numpy as np
import warnings 
warnings.filterwarnings('ignore')
```

```python
df=pd.DataFrame({
    'Name':['Aman','Bob','chandan','david','eva','freddy','gagan','hugh','imad','jack'],
    'Age':[22,23,24,34,45,35,27,26,29,30],
    'city':['amravati','nagpur','wardha','asegaon','akola','washim','bhusawal','bangaluru','mumbai','chennai'],
    'salary':[35000,40000,45000,np.nan,50000,55000,np.nan,60000,65000,70000]
})
```

```python
df.info()
```

```python
df.describe()
```

```python
df[['Name','city']]
```

```python
df.drop('city', axis=1, inplace=True)
```

```python
df.duplicated().isnull()
```

```python
df['salary'].fillna(df['salary'].mean(), inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```
